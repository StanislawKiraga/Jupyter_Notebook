```python
import pandas as pd
import matplotlib.pyplot as plt

prices = [
	(1, 2.12),
	(2, 2.56),
	(3, 3.10),
	(4, 3.16),
	(5, 3.58),
	(6, 5.12),
	(7, 5.16),
	(8, 5.20),
	(9, 4.12),
	(10, 4.10),
	(11, 3.65),
	(12, 4.25),

]

df = pd.DataFrame(prices, columns=['month', 'price'])
df['priceUSD'] = [round((pr / 4), 2) for pr in df['price']]
df = df.set_index('month')

df['priceUSD'].plot(kind='line', title='Price of goods(USD)', color='#FF0606', linestyle='dashed', grid=True).axis([0,12.5,0,2])
```
